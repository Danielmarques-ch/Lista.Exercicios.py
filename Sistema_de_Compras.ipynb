```python
# Listas
produtos = ['Celular', 'Fone', 'Notebook', 'Monitor']
estoque = [1200, 3000, 4000, 6000]
precos = [1500, 200, 4500, 900]

# Mostra catálogo (FOR)
print('--- Catálogo de Produtos ---')
for i in range(len(produtos)):
    print(f'{produtos[i]} | Preço: R$ {precos[i]} | Estoque: {estoque[i]}')

# Produto escolhido (WHILE)
produto = input('\nInforme qual produto deseja: ')

while produto not in produtos:
    print('Produto inválido')
    produto = input('Informe o produto desejado: ')

# Descobre a posição do produto (FOR)
for i in range(len(produtos)):
    if produtos[i] == produto:
        indice = i
        break

# Mostra informações do produto escolhido
print(f'\nQuantidade disponível de {produto}: {estoque[indice]} unidades')
print('Esse produto está no valor de R$ {:.0f}'.format(precos[indice]))

# Quantidade que o cliente quer comprar (WHILE)
qtd_compra = int(input('Informe a quantidade do produto que deseja adquirir: '))

while qtd_compra > estoque[indice]:
    print('Quantidade indisponível em estoque')
    qtd_compra = int(input('Informe uma quantidade válida: '))

# Calcula o valor final
price_final = qtd_compra * precos[indice]
print(f'Valor total da compra: R$ {price_final:.2f}')

# Confirmação final (WHILE)
validacao = input('Digite "SIM" para continuar sua compra ou "NÃO" para finalizar o atendimento: ')

while validacao != 'SIM' and validacao != 'NÃO':
    validacao = input('Digite apenas "SIM" ou "NÃO": ')

if validacao == 'SIM':
    print('Daremos continuidade')
else:
    print('Agradecemos a utilização da nossa plataforma, volte sempre!')

# Forma de pagamento
forma_pagamento = input('Forma de pagamento (PIX / CARTÃO / BOLETO): ')

while forma_pagamento not in ['PIX', 'CARTÃO', 'BOLETO']:
    forma_pagamento = input('Informe PIX, CARTÃO ou BOLETO: ')

print(f'Pagamento via {forma_pagamento} confirmado.')

# Desconto na forma de pagamento
if forma_pagamento == 'PIX':
    desconto = price_final * 0.05
    price_final -= desconto
    print(f'Desconto aplicado: R$ {desconto:.2f}')

# Questiona se deseja comprar outro produto
continuar = input('Deseja comprar outro produto? (SIM / NÃO): ')

while continuar == 'SIM':
    print('\nVoltando ao catálogo...')
    # aqui você poderia repetir o fluxo
    continuar = input('Deseja comprar outro produto? (SIM / NÃO): ')
vendas = []

vendas.append({
    'produto': produto,
    'quantidade': qtd_compra,
    'total': price_final
})

print('Venda registrada com sucesso.')

```
