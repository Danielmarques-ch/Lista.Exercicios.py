```python
# Listas para armazenar os candidatos
nomes = []
idades = []
areas = []

# Áreas disponíveis
setores = ['Dados', 'Administrativo', 'TI']

# Controle do sistema
continuar = 's'

while continuar == 's':

    print('\n--- NOVO CADASTRO DE CANDIDATO ---')

    nome_candidato = input('Informe seu nome completo: ')
    idade = int(input('Informe sua idade: '))

    # Validação de idade
    if idade < 16:
        print('Você não pode completar o cadastro devido à sua idade.')
    else:
        print('Temos as seguintes vagas disponíveis: Dados, Administrativo, TI')

        area = input('Informe a área desejada: ').strip().title()
        while area not in setores:
            print('Área informada não está na nossa base.')
            area = input('Informe novamente a área desejada: ').strip().title()

        # Salvando os dados
        nomes.append(nome_candidato)
        idades.append(idade)
        areas.append(area)

        print(f'''
Cadastro realizado com sucesso!
Nome: {nome_candidato}
Idade: {idade}
Área: {area}
''')

    # Pergunta se deseja continuar
    continuar = input('Deseja cadastrar outro candidato? (s/n): ').strip().lower()
# Total de candidatos
total = len(nomes)

# Contadores por área
dados = 0
administrativo = 0
ti = 0

for area in areas:
    if area == 'Dados':
        dados += 1
    elif area == 'Administrativo':
        administrativo += 1
    elif area == 'TI':
        ti += 1

print('\n--- RESUMO DO PROCESSO SELETIVO ---')
print(f'Total de candidatos: {total}')
print(f'Dados: {dados} candidatos')
print(f'Administrativo: {administrativo} candidatos')
print(f'TI: {ti} candidatos')


```
