```python
import streamlit as st
import pandas as pd

# Lista de bairros de Cariacica/ES
bairros_cariacica = [
    "Alto Lage", "Alzira Ramos", "Amar√≠lio Caiado", "Bananal", "Bela Aurora",
    "Boa Sorte", "Campo Verde", "Cascata", "Cristo Rei", "Flexal I", "Flexal II",
    "Itacib√°", "Itangu√°", "Jardim Am√©rica", "Maracan√£", "Mucuri", "Nova Bras√≠lia",
    "Nova Campo Grande", "Nova Rosa da Penha I", "Nova Rosa da Penha II",
    "Nova Valverde", "Oriente", "Porto de Santana", "Prolar", "Rio Marinho",
    "Santa B√°rbara", "Santa Cec√≠lia", "Santa F√©", "Santa Luzia", "Santa Paula I",
    "Santa Paula II", "Santa Rosa", "S√£o Benedito", "S√£o Conrado", "S√£o Francisco",
    "S√£o Geraldo", "S√£o Jo√£o Batista", "S√£o Jorge", "S√£o Pedro", "S√£o Torquato",
    "Vale Esperan√ßa", "Vila Capixaba", "Vila Isabel", "Vila Merlo", "Vista Dourada"
]

# T√≠tulo
st.title("üì¶ Filtro de Rotas ‚Äì Escolha Bairros (Cariacica/ES)")

# Multisele√ß√£o de bairros
bairros_selecionados = st.multiselect(
    "Selecione os bairros desejados:",
    bairros_cariacica
)

# Upload da planilha
uploaded_file = st.file_uploader("Envie a planilha de rotas (Excel)", type=["xlsx", "xls"])

if uploaded_file:
    try:
        # L√™ a planilha
        df = pd.read_excel(uploaded_file)

        # Padroniza os nomes das colunas para min√∫sculo
        df.columns = df.columns.str.lower()

        # Verifica se existe a coluna "bairro"
        if "bairro" not in df.columns:
            st.error("‚ùå A planilha n√£o cont√©m a coluna 'bairro'. Verifique o cabe√ßalho.")
        else:
            # Filtra os bairros selecionados
            if bairros_selecionados:
                df_filtrado = df[df["bairro"].isin(bairros_selecionados)]
                st.success(f"‚úÖ {len(df_filtrado)} registros encontrados nos bairros selecionados.")
                st.dataframe(df_filtrado)

                # Bot√£o para download
                csv = df_filtrado.to_csv(index=False).encode("utf-8-sig")
                st.download_button(
                    label="üì• Baixar planilha filtrada",
                    data=csv,
                    file_name="rotas_filtradas.csv",
                    mime="text/csv",
                )
            else:
                st.warning("‚ö†Ô∏è Selecione pelo menos um bairro para aplicar o filtro.")

    except Exception as e:
        st.error(f"Erro ao processar a planilha: {e}")
```
