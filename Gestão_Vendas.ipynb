```python
# ==========================================
# SISTEMA GERENCIAL DE VENDAS - PYTHON
# Autor: Daniel (sim, isso aqui é projeto real)
# ==========================================

import matplotlib.pyplot as plt  # Biblioteca para gráficos (funciona no Colab)

# ------------------------------
# CONFIGURAÇÕES INICIAIS
# ------------------------------

meta = 25000  # Meta de vendas

nomes_vendedores = []
valores_vendas = []

# ------------------------------
# FUNÇÃO PARA CADASTRAR VENDEDORES
# ------------------------------

def cadastrar_vendedores():
    while True:
        vendedor = input('Informe o nome do vendedor: ')

        # Validação de nome vazio
        while vendedor == '':
            print('O nome não pode ficar vazio.')
            vendedor = input('Informe o nome do vendedor: ')

        venda = float(input('Informe o valor vendido: '))

        nomes_vendedores.append(vendedor)
        valores_vendas.append(venda)

        continuar = input('Deseja cadastrar outro vendedor? (s/n): ').lower()
        if continuar != 's':
            break

# ------------------------------
# FUNÇÃO PARA GERAR RELATÓRIO
# ------------------------------

def gerar_relatorio():
    total_vendas = sum(valores_vendas)
    media_vendas = total_vendas / len(valores_vendas)

    print('\n========== RELATÓRIO DE VENDAS ==========')
    print(f'Total de vendedores: {len(nomes_vendedores)}')
    print(f'Total vendido: R$ {total_vendas:.2f}')
    print(f'Média por vendedor: R$ {media_vendas:.2f}')

    print('\n--- DESEMPENHO INDIVIDUAL ---')

    for i in range(len(nomes_vendedores)):
        percentual = (valores_vendas[i] / meta) * 100
        falta_percentual = max(0, 100 - percentual)
        falta_reais = max(0, meta - valores_vendas[i])

        if percentual < 100:
            print(
                f'{nomes_vendedores[i]} vendeu R$ {valores_vendas[i]:.2f} | '
                f'Atingiu {percentual:.1f}% da meta | '
                f'Falta {falta_percentual:.1f}% (R$ {falta_reais:.2f})'
            )
        else:
            print(
                f'{nomes_vendedores[i]} vendeu R$ {valores_vendas[i]:.2f} | '
                f'Superou a meta ({percentual:.1f}%)'
            )

    print('========== FIM DO RELATÓRIO ==========')

# ------------------------------
# FUNÇÃO PARA RANKING DE VENDEDORES
# ------------------------------

def ranking_vendedores():
    print('\n========== RANKING DE VENDAS ==========')

    # Junta nome e venda e ordena do maior para o menor
    ranking = list(zip(nomes_vendedores, valores_vendas))
    ranking.sort(key=lambda x: x[1], reverse=True)

    for posicao, (nome, venda) in enumerate(ranking, start=1):
        print(f'{posicao}º - {nome}: R$ {venda:.2f}')

# ------------------------------
# FUNÇÃO PARA GRÁFICO
# ------------------------------

def grafico_vendas():
    plt.figure()
    plt.bar(nomes_vendedores, valores_vendas)
    plt.axhline(meta)  # Linha da meta
    plt.title('Vendas por Vendedor')
    plt.xlabel('Vendedores')
    plt.ylabel('Valor Vendido (R$)')
    plt.show()

# ------------------------------
# MENU INTERATIVO
# ------------------------------

while True:
    print('\n===== MENU =====')
    print('1 - Cadastrar vendedores')
    print('2 - Gerar relatório')
    print('3 - Mostrar ranking')
    print('4 - Mostrar gráfico de vendas')
    print('0 - Sair')

    opcao = input('Escolha uma opção: ')

    if opcao == '1':
        cadastrar_vendedores()

    elif opcao == '2':
        if nomes_vendedores:
            gerar_relatorio()
        else:
            print('Nenhum vendedor cadastrado.')

    elif opcao == '3':
        if nomes_vendedores:
            ranking_vendedores()
        else:
            print('Nenhum vendedor cadastrado.')

    elif opcao == '4':
        if nomes_vendedores:
            grafico_vendas()
        else:
            print('Nenhum vendedor cadastrado.')

    elif opcao == '0':
        print('Sistema encerrado.')
        break

    else:
        print('Opção inválida. Tente novamente.')
```
